```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

dataframe= pd.read_csv('salary_data.csv')

dataframe=np.array(dataframe)
X=dataframe[:,:-1]
y=dataframe[:,-1:]

regressor = LinearRegression()
#regressor.fit(X,y)

XTrain,XTest,yTrain,yTest=train_test_split(X,y,test_size=0.3,random_state=0)

regressor.fit(XTrain,yTrain)
#print (XTest)
regressor.predict(XTest)
regressor.score(XTest,yTest)
regressor.score(XTrain,yTrain)
regressor.predict([[9.2]])

plt.scatter(XTrain,yTrain,color="red",label="train")
plt.scatter(XTest,yTest,color="blue",label="test")

plt.plot(XTest,regressor.predict(XTest),color="green",label="regression line")
plt.legend()

print(regressor.coef_)
print(regressor.intercept_)
```
